```python
#Import Libraries
import numpy as np
import nltk
import pandas as pd
import re
from textblob import TextBlob
from sklearn import preprocessing
```

```python
#Read Dataset
data1 = pd.read_csv('Hotelscombined_com.csv',encoding='latin1')
data1
```

```python
#Normalize Ratings
r1 = data1['Hotel_ratings']
r1 = r1.str.extract('(\d+)')
min_max_scaler = preprocessing.MinMaxScaler()
r1_scaled = min_max_scaler.fit_transform(r1)
```

```python
#Reviews
re1 = data1['Review']
```

```python
#Lexicon
stopwords = [ "hmm","[","]",'·','·',"...","ok","th","rd","th","nd","!","#","$","%","&","'","’","u","v","b","r","(",")","*","+","-",".","/",":",";","<","=",">","?","@","[","/","]","^","_","`","{","|","}","~",",","a", "about", "above", "after", "again", "against", "all", "am", "an", "and", "any", "are", "as", "at", "be", "because", "been", "before", "being", "below", "between", "both", "but", "by", "could", "did", "do", "does", "doing", "down", "during", "each", "few", "for", "from", "further", "had", "has", "have", "having", "he", "he'd", "he'll", "he's", "her", "here", "here's", "hers", "herself", "him", "himself", "his", "how", "how's", "i", "i'd", "i'll", "i'm", "i've", "if", "in", "into", "is", "it", "it's", "its", "itself", "let's", "me", "more", "most", "my", "myself", "nor", "of", "on", "once", "only", "or", "other", "ought", "our", "ours", "ourselves", "out", "over", "own", "same", "she","please","sorry","bye", "she'd", "she'll", "she's", "should", "so", "some", "such", "than", "that", "that's", "the", "their", "theirs", "them", "themselves", "then", "there", "there's", "these", "they", "they'd", "they'll", "they're", "they've", "this", "those", "through", "to", "too", "under", "until", "up", "very", "was", "we", "we'd", "we'll", "we're", "we've", "were", "what", "what's", "when", "when's", "where", "where's", "which", "while", "who", "who's", "whom", "why", "why's", 'nan',"with", "would", "you", "you'd", "you'll", "you're", "you've", "your", "yours", "yourself", "yourselves","hello"]
```

```python
#Data Structures
Hotel_neg = []
Hotel_pos = []
Hotel_neu = []
for re in re1:
    re = str(re)
    
    #Lower Case
    re = re.lower()
    
    #Word Tokenization
    re = nltk.word_tokenize(re)
    
    #POS Tagger
    tags3 = nltk.pos_tag(re)
    R=[]
    N=[]
    P=[]
    for t in tags3:
            log = (t[1] == 'NN')
            if log == True:
                R.append(t[0])
                
    #Stopword Removal
    R = [words for words in R if words not in stopwords]
    
    #Sentimental Analysis
    for x in R:
        s = TextBlob(x)
        if s.sentiment.polarity < 0:
            N.append(x)
    for x in R:
        s = TextBlob(x)
        if s.sentiment.polarity > 0:
            P.append(x)
            
    #Counter
    #Top most positive features
    #Top most negative features
    #Top most talked about features
    from collections import Counter 
    Counter1 = Counter(N) 
    Counter2 = Counter(P) 
    Counter3 = Counter(R) 
    
    #Most Common Words
    most_occur_n = Counter1.most_common(5) 
    most_occur_p = Counter2.most_common(5)
    most_occur_r = Counter3.most_common(5)
    
    #Final Columns 
    Hotel_neg.append(str(most_occur_n))
    Hotel_pos.append(str(most_occur_p))
    Hotel_neu.append(str(most_occur_r))
```

```python
#Data Structures
Com= []
Diff = []
Comx=[]
Diffx= []

#Hotel Cities
Cities = data2['Hotel_city']
Cities = Cities.unique()

for c in Cities:
    
    #Hotels of each city 
    data = data2[data2['Hotel_city'].str.contains(c)] 
    c = c.lower()
    data = data.append(data2[data2['Hotel_city'].str.contains(c)], ignore_index=True )
    
    #Hotel names
    name = data['Hotel_name']
    
    com = []
    dif = []
    
    for n in name:
        
        #Hotel facilities/features
        datas = data[data['Hotel_name']!=n]
        fac = datas['Hotel_facilities']
        fn = data[data['Hotel_name']==n]
        fn = fn['Hotel_facilities']
        fn = fn.iloc[0]
        fn = str(fn)
        fn = fn.lower()
        fn = fn.lower()
        fn = fn.split(',')
        fn = list(fn)
        Facs = []
        for f in fac:
            f = str(f)
            f = f.lower()
            f = f.split(',')
            Facs.append(f)
        Facs = np.array(Facs)
        Facs = np.unique(Facs)
        
        #Common Features with surrounding hotels
        com = [words for words in Facs if words in fn ]
        
        #Different Features from surrounding hotels
        diff = [words for words in Facs if words not in fn ]
        s = []
        w = []
        for g in fn:
            for k in Facs:
                    for l in k:
                        if g==l:
                            if g not in s:
                                s.append(g)
        for g in fn:
            t=0
            for k in Facs:
                for l in k:
                    if g==l:
                        t=1
            if t!=1:
                w.append(g)
        s = str(s)
        w = str(w)
        
        #Final Columns 
        Comx.append(s)
        Diffx.append(w)      
        Com.append(com)
        Diff.append(diff)
```

```python
#DataFrame Creation
p = pd.DataFrame(Diffs)
q = pd.DataFrame(r1_scaled)
r = pd.DataFrame(Coms)
s=pd.DataFrame(r1_scaled)
t=pd.DataFrame(Hotel_neg)
u=pd.DataFrame(Hotel_neu)
v=pd.DataFrame(Hotel_pos)
Data1 = pd.concat([q,s,v,t,u,r,p],axis=1)
Data1
```
