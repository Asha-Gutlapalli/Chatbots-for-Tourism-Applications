```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def parser(url):   
    data = requests.get(url)
    soup = BeautifulSoup(data.text, 'html.parser')
    return soup
```

```python
sp = parser('https://in.hotels.com/search.do?resolved-location=CITY%3A1640701%3AUNKNOWN%3AUNKNOWN&destination-id=1640701&q-destination=Mumbai,%20Maharashtra,%20India&q-check-in=2020-01-18&q-check-out=2020-01-19&q-rooms=1&q-room-0-adults=2&q-room-0-children=0')
```

```python
Hotel_name = []
Hotel_badge = []
Address = []
Hotel_id = []
Amenities = []
Ratings = []
No_of_reviews = []
Price = []
hotels = sp.find_all('li',{'class':'hotel'})
for hotel in hotels:
    hotel_name = hotel.a.text.strip()
    hotel_badge = hotel.span.text
    address = hotel.address.text
    hotel_id = hotel['data-hotel-id']
    amenities = hotel.find('ul',{'class':'hmvt8258-amenities'})
    if amenities:
        amenities = amenities.text
    else:
        amenities = 'None'
    ratings = hotel.strong.text
    no_of_reviews = hotel.find('span',{'class':'small-view'})
    if no_of_reviews:
        no_of_reviews = no_of_reviews.text
    else:
        no_of_reviews = 'None'
    price = hotel.ins
    if price:
        price = price.text
    else:
        price = 'None'
    Hotel_name.append(hotel_name)
    Hotel_badge.append(hotel_badge)
    Address.append(address)
    Hotel_id.append(hotel_id)
    Amenities.append(amenities)
    Ratings.append(ratings)
    No_of_reviews.append(no_of_reviews)
    Price.append(price)
```

```python
Hotels = pd.DataFrame()
Hotels['Hotel_id'] = Hotel_id
Hotels['Hotel_name'] = Hotel_name
Hotels['Hotel_badge'] = Hotel_badge
Hotels['Address'] = Address
Hotels['Amenities'] = Amenities
Hotels['No_of_reviews'] = No_of_reviews
Hotels['Ratings'] = Ratings
Hotels['Price'] = Price
```
