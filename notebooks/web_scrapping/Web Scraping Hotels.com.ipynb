```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def parser(url):   
    data = requests.get(url)
    soup = BeautifulSoup(data.text, 'lxml')
    return soup
```

```python
sp = parser('https://www.hotelscombined.com/Place/Dubai_1.htm')
```

```python
d = sp.find_all('a')
h = sp.find_all('h3')
```

```python
expan = sp.find_all('li',{'class':'hotel expanded-area'})
for hotel in expan:
    hotel_name = hotel.a.text.strip()
    hotel_badge = hotel.span.text
    address = hotel.address.text
    hotel_id = hotel['data-hotel-id']
    amenities = hotel.find('ul',{'class':'hmvt8258-amenities'}).text
    ratings = hotel.strong.text
    no_of_reviews = hotel.find('span',{'class':'small-view'}).text
    hotel_price = hotel.find('div',{'class':'price'})
    price = hotel_price.strong
    if price:
        price = price.text
    else:
        price = hotel.ins.text
    Hotel_name.append(hotel_name)
    Hotel_badge.append(hotel_badge)
    Address.append(address)
    Hotel_id.append(hotel_id)
    Amenities.append(amenities)
    Ratings.append(ratings)
    No_of_reviews.append(no_of_reviews)
    Price.append(price)
```

```python
secret = sp.find_all('li',{'class':'hotel secret-price-badge-available'})
for hotel in secret:
    hotel_name = hotel.a.text.strip()
    hotel_badge = hotel.span.text
    address = hotel.address.text
    hotel_id = hotel['data-hotel-id']
    amenities = hotel.find('ul',{'class':'hmvt8258-amenities'})
    if amenities:
        amenities = amenities.text
    else:
        amenities = 'None'
    ratings = hotel.strong.text
    no_of_reviews = hotel.find('span',{'class':'small-view'})
    if no_of_reviews:
        no_of_reviews = no_of_reviews.text
    else:
        no_of_reviews = 'None'
    hotel_price = hotel.find('div',{'class':'price'})
    price = hotel_price.strong
    if price:
        price = price.text
    else:
        price = hotel.ins.text
    Hotel_name.append(hotel_name)
    Hotel_badge.append(hotel_badge)
    Address.append(address)
    Hotel_id.append(hotel_id)
    Amenities.append(amenities)
    Ratings.append(ratings)
    No_of_reviews.append(no_of_reviews)
    Price.append(price)
```

```python
hotels = sp.find_all('li',{'class':'hotel'})
for hotel in hotels:
    hotel_name = hotel.a.text.strip()
    hotel_badge = hotel.span.text
    address = hotel.address.text
    hotel_id = hotel['data-hotel-id']
    amenities = hotel.find('ul',{'class':'hmvt8258-amenities'})
    if amenities:
        amenities = amenities.text
    else:
        amenities = 'None'
    ratings = hotel.strong.text
    no_of_reviews = hotel.find('span',{'class':'small-view'})
    if no_of_reviews:
        no_of_reviews = no_of_reviews.text
    else:
        no_of_reviews = 'None'
    price = hotel.ins
    if price:
        price = price.text
    else:
        price = 'None'
    Hotel_name.append(hotel_name)
    Hotel_badge.append(hotel_badge)
    Address.append(address)
    Hotel_id.append(hotel_id)
    Amenities.append(amenities)
    Ratings.append(ratings)
    No_of_reviews.append(no_of_reviews)
    Price.append(price)
```

```python
Hotels_com = pd.DataFrame()
Hotels_com['Hotel_id'] = Hotel_id
Hotels_com['Hotel_name'] = Hotel_name
Hotels_com['Hotel_badge'] = Hotel_badge
Hotels_com['Address'] = Address
Hotels_com['Amenities'] = Amenities
Hotels_com['No_of_reviews'] = No_of_reviews
Hotels_com['Ratings'] = Ratings
Hotels_com['Price'] = Price
```
